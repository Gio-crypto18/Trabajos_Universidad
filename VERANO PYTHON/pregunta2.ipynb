```python
import pandas as pd
import numpy as np
import numpy as np
from matplotlib import pyplot as plt

class DataFrame:
    def __init__(self):
        URL = ("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
        self.data = pd.read_csv(URL, header=None)
        self.columnas = ["sepallength", "sepalwidth", "petallength", "petalwidth", "species"]
    

    def Basico(self):
        print(f"Los 10 primeros resultados\n{self.data.head(10)}")
        print(f"Los 10 ultimos resultados\n{self.data.tail(10)}")


    def ponerColumnas(self):
        self.data.columns = self.columnas
        print(f"Los 10 primeros resultados con encabezado\n{self.data.head(10)}")
        print(f"Los 10 ultimos resultados con encabezado\n{self.data.tail(10)}")


    def Calculo(self):
        media = round(self.data["sepallength"].mean(), 4)
        print(f"\nLa media de sepallength es {media} ")
        mediana = self.data["sepallength"].median()
        print(f"\nLa mediana de sepallength es {mediana} ")
        moda = self.data["sepallength"].mode()
        print(f"\nLa moda de sepallength es \n{moda} ")
        max = self.data["sepallength"].max()
        print(f"\nEl valor maximo de sepallength es \n{max} ")


    def AscendenteyDescendente(self):
        print(self.data.sort_values(["sepalwidth", "petallength"], ascending=[True, False]))


    def DiagramaBarras(self):

        petalwitdh = np.array(self.data.loc[:, "petalwidth"])
        petalwitdh1 = np.array(self.data.loc[:, "petalwidth"])

        plt.bar(petalwitdh, petalwitdh, color='#33FF36', label='Cantidad', linewidth=3)
        plt.bar(petalwitdh1, petalwitdh1, color='#3D33FF', label='Cantdad', linewidth=2)
        plt.title('Cantidad')
        plt.ylabel('Cantidad')
        plt.xlabel('Nmeros')
        plt.legend()
        plt.grid(color='#FF334E')
        plt.show()

    def Resumen(self):
        species = np.array(self.data.loc[:, "species"])
        print(species)


    def DiagramaPie(self):
        species = np.array(self.data.loc[:, "species"])
        colores = ["#EE6055","#60D394","#AAF683","#FFD97D","#FF9B85"]
        plt.pie(species, autopct="%0.1f%%", colors=colores)
        plt.show()

    def CopiayContaryEliminar(self):
        self.dataCopia = self.data.copy()
        x = np.random.randint(3000, size=30)
        y = np.random.randint(9, size=30)
        for i in range(30):
            self.dataCopia.loc[x[i], self.columnas[y[i]]] = np.nan
        print(f"Con Nan insertamos en el dataframe\n{self.dataCopia.head(30)}")
        print(self.dataCopia.isnull().sum())
        self.dataCopia = self.dataCopia.dropna()
        print(f"Data sin Nan\n{self.dataCopia.head(50)}")
        print(f"Data sin Nan\n{self.dataCopia.tail(50)}")


Data = DataFrame()
Data.Basico()
Data.ponerColumnas()
Data.Calculo()
Data.AscendenteyDescendente()
Data.DiagramaBarras()
Data.Resumen()
Data.DiagramaPie()
Data.menu()
Data.CopiayContaryEliminar()

def menu(self):
    while True:
        print("Menu ")
        print("-------------------------------------------------------")
        print("1.Lectura de datos")
        print("2.Encabezados" 
              "2017")
        print("3.Calculos sobre el dataframe" 
              "de Cajamarca?")
        print("4.Acendente y descendente")
        print("5.Diagrama de barras?")
        print("6.Diagrama pie")
        print("0.Salir")
        try:
            opcion = int(input("Que reporte se desea visualizar?: "))
            if opcion == 0:
                print("Programa finalizado")
                break
            elif opcion == 1:
                print(f"\n{Data.Basico()}\n\n")
            elif opcion == 2:
                print(f"\n{Data.ponerColumnas()}\n\n")
            elif opcion == 3:
                print(f"\n{Data.Calculo()}\n\n")
            elif opcion == 4:
                print(f"\n{Data.AscendenteyDescendente()}\n\n")
            elif opcion == 5:
                print(f"\n{Data.DiagramaBarras()}\n\n")
            elif opcion == 6:
                print(f"\n{Data.DiagramaPie()}\n\n")
            else:
                print("ERROR ingrese una de las opciones")
        except ValueError:
            print("ERROR ingrese un nmero")
```
